```python
import pandas as pd

file_path = "data/final_merged.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Filter the dataframe for matches between Duke (team id 1181) and Florida (team id 1196)
duke_florida_matches = df[((df['WTeamID'] == 1181) & (df['LTeamID'] == 1196)) | ((df['WTeamID'] == 1196) & (df['LTeamID'] == 1181))]

# Calculate the average score for each team
average_score_duke = (duke_florida_matches[duke_florida_matches['WTeamID'] == 1181]['WScore'].mean() + duke_florida_matches[duke_florida_matches['LTeamID'] == 1181]['LScore'].mean()) / 2
average_score_florida = (duke_florida_matches[duke_florida_matches['WTeamID'] == 1196]['WScore'].mean() + duke_florida_matches[duke_florida_matches['LTeamID'] == 1196]['LScore'].mean()) / 2

# Replace NaN values with 0
average_score_duke = average_score_duke if not pd.isna(average_score_duke) else 0
average_score_florida = average_score_florida if not pd.isna(average_score_florida) else 0

print(f"Average score of Duke against Florida: {average_score_duke}")
print(f"Average score of Florida against Duke: {average_score_florida}")
```

```python
# Filter the dataframe for matches between Texas (team id 2563) and Connecticut (team id 2588) in women's games
texas_connecticut_matches = df[((df['WTeamID'] == 2563) & (df['LTeamID'] == 2588)) | ((df['WTeamID'] == 2588) & (df['LTeamID'] == 2563))]

# Calculate the average score for each team
average_score_texas = (texas_connecticut_matches[texas_connecticut_matches['WTeamID'] == 2563]['WScore'].mean() + texas_connecticut_matches[texas_connecticut_matches['LTeamID'] == 2563]['LScore'].mean()) / 2
average_score_connecticut = (texas_connecticut_matches[texas_connecticut_matches['WTeamID'] == 2588]['WScore'].mean() + texas_connecticut_matches[texas_connecticut_matches['LTeamID'] == 2588]['LScore'].mean()) / 2

# Replace NaN values with 0
average_score_texas = average_score_texas if not pd.isna(average_score_texas) else 0
average_score_connecticut = average_score_connecticut if not pd.isna(average_score_connecticut) else 0

print(f"Average score of Texas against Connecticut: {average_score_texas}")
print(f"Average score of Connecticut against Texas: {average_score_connecticut}")
```
